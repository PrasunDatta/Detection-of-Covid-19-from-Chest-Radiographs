```
CovidCoeff =[]
# pneumoniaCoeff =[]
# imagesall = glob.glob("/content/drive/My Drive/Abar_ dataset/covid/*")
testall = glob.glob('/content/drive/My Drive/Abar_ dataset/test_covid/*')

for k in range(len(testall)):
    image = cv2.imread("/content/drive/My Drive/Abar_ dataset/test_covid/"+"test_covid_{}.jpg".format(k+1))
    # image = cv2.imread('/content/drive/My Drive/Abar_ dataset/covid/' + 'covid_{}.jpg'.format(k+1))
    # print('/content/drive/My Drive/Abar_ dataset/covid/' + 'covid_{}.jpg'.format(k+1))
    print("/content/drive/My Drive/Abar_ dataset/test_covid/"+"test_covid_{}.jpg".format(k+1))
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image,(224,224))
    image = np.float32(image)
    image /= 255
    coeffs2 = pywt.dwt2(image, 'haar')
    LL, (LH, HL, HH) = coeffs2
    LLflatten = LL.flatten()
    LHflatten = LH.flatten()
    HLflatten = HL.flatten()
    HHflatten = HH.flatten()
    total  = np.append(LLflatten,LHflatten)
    total1 = np.append(total ,HLflatten )
    total2 = np.append(total1,HHflatten)    
    # CovidCoeff.append(total2)
    CovidCoeff.append(total2)
# Coviddf = pd.DataFrame(CovidCoeff)
# Coviddf.to_csv('/content/drive/My Drive/(Prasun)AllCodes&Results/FinPreCov - Sheet1.csv', index =False)    
```
